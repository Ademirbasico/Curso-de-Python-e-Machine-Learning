10. Exercícios Propostos
Crie um programa que solicite o nome e a idade do usuário e exiba uma mensagem personalizada.
Leia dois números digitados pelo usuário e exiba a soma.
Calcule a média de três notas informadas pelo usuário.

```python
nome = input("Insira seu nome: ")
idade = int(input("Insira sua idade: "))

nota = int(input("Digite sua primeira nota: "))
nota2 = int(input("Digite sua segunda nota: "))
nota3 = int(input("Digite sua terceira nota:"))

soma = nota + nota2 + nota3
media_final = soma/3

print("Seu Nome:"  , nome, "\n Sua idade:  ", idade,"\n A soma das suas notas foram: ",soma, "\n Sua media final: ", media_final)
```
